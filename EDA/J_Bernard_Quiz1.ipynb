# EDA- diamonds.csv

```python
import numpy as np
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as mpl
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jonpb99/Data_Analytics_List_Comprehension/main/EDA/diamonds.csv')
```

```python
ndf = df.T.to_numpy()
price = ndf[7]
carat = ndf[1]
cut = ndf[2]
color = ndf[3]
clarity = ndf[4]
x = ndf[8]
y = ndf[9]
z = ndf[10]
depth = ndf[5]
table = ndf[6]
```

diamonds.csv is a dataset of 53940 entries, with 10 attibutes each
 
Price entered as int values

Carat, x(length), y(width), z(depth), depth(total depth percentage)
table(width of top compared to widest point), and carat(weight)
entered as float values.
cut, color, clarity entered as object values that 
appear like string values; all are graded values
